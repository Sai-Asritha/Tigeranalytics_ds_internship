```python
#Q1toQ6
import pandas as pd
```

```python
sales=pd.read_excel("/SalesData.xlsx")
sales.head()
```

```python
#1
least_amount_sale= sales.groupby("Item")["Sale_amt"].min().reset_index()
least_amount_sale
```

```python
#2
sales["Year"]=sales["OrderDate"].dt.year
totalsales_yearregion= sales.groupby(["Year", "Region"])["Sale_amt"].sum().reset_index()
totalsales_yearregion
```

```python
#3
from datetime import datetime
inputyear=int(input("enter year"))
inputmonth=int(input("enter month"))
inputdate=int(input("enter date"))
reference_date=datetime(inputyear,inputmonth,inputdate)
sales["days_diff"]=(reference_date-sales["OrderDate"]).dt.days
print(sales[["OrderDate","days_diff"]])

```

```python
#4
salesman_manager=sales.groupby("Manager")["SalesMan"].unique().reset_index()
salesman_manager
```

```python
#5
sales_info=sales.groupby("Region").agg(
    total_sales=("Sale_amt","sum"),
    total_salesman=("SalesMan","count",)).reset_index()
sales_info


```

```python
#6
total=sales["Sale_amt"].sum()
percent_sales=sales.groupby("Manager")["Sale_amt"].sum()/total*100
percent_sales.reset_index()
```
